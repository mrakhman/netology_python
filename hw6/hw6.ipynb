### Задание 1

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток. Результат классификации запишите в отдельный столбец region.

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
import pandas as pd
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def test(request, geo_data):
    request = request.lower().strip().split(' ')
    for word in request:
        if word in geo_data['Центр']:
            return 'Центр'
        if word in geo_data['Северо-Запад']:
            return 'Северо-Запад'
        if word in geo_data['Дальний Восток']:
            return 'Дальний Восток'
    return 'undefined'

x = test('this city is ПетербурГ', geo_data)
y = test('владивосток this city is ', geo_data)
z = test('владивостоке', geo_data)
print(x, y, z)

""" 
Этот способ норм, но не работает с падежами слова. 
Но тогда надо редактировать и geo_data, так что наверное это не требуется
""" 
```

```python
def region(row, geo_data):
    request = row['keyword']
    request = str(request).lower().strip().split(' ')
    for word in request:
        if word in geo_data['Центр']:
            return 'Центр'
        if word in geo_data['Северо-Запад']:
            return 'Северо-Запад'
        if word in geo_data['Дальний Восток']:
            return 'Дальний Восток'
    return 'undefined'
```

```python
keywords['region'] = keywords.apply(region, geo_data=geo_data, axis=1)
```

### Задание 2

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('ratings.csv')
```

```python
avg_rating = ratings.groupby('movieId').mean()[['rating']]
avg_rating.head()
```

```python
def rating_class(row):
    level = row['rating']
    if level <= 2:
        return 'Low'
    if level <= 4:
        return 'Medium'
    if level <= 5:
        return 'High'
    return 'Mistake'
```

```python
avg_rating['class'] = avg_rating.apply(rating_class, axis=1)
```

### Задание 3

Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого пользователя. Ответ дайте в днях.

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
maniacs = ratings.groupby('userId').count()[['rating']]
maniacs = maniacs[maniacs['rating'] > 100].reset_index()
maniacs.head()
```

```python
maniacs_list = maniacs['userId'].tolist()
maniacs_list
```

```python
ratings_maniacs = ratings[ratings['userId'].isin(maniacs_list)].groupby('userId').agg(['min', 'max'])['timestamp']
ratings_maniacs.head()
```

```python
from datetime import datetime

def lifetime(row):
    new_t = row['max']
    old_t = row['min']
    n_days = (new_t - old_t) / 86400
    return(n_days)
```

```python
# Ответ

ratings_maniacs['lifetime'] = ratings_maniacs.apply(lifetime, axis=1)
ratings_maniacs.head()
```

### Задание 4

Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    - для каждой строки пройдите по всем годам списка years
    - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

```python
years = list(range(1950, 2011))
# years
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
def production_year(row, years):
    film_title = row['title']
    for year in years:
        if str(year) in film_title:
            return year
    return 1900
```

```python
movies['year'] = movies.apply(production_year, years=years, axis=1)
```

```python
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
avg_rating = ratings.groupby('movieId').mean()[['rating']]
```

```python
average_rating = movies.merge(avg_rating, on='movieId', how='left')
average_rating.head()
```

```python
rating_by_year = average_rating.groupby('year').mean()[['rating']].sort_values(by='rating', ascending=False)
rating_by_year.head(60)
```

```python
rating_by_year['rating'].max() - rating_by_year['rating'].min()
```
